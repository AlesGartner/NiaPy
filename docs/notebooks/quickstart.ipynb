```python
import random
from NiaPy.algorithms.basic import GreyWolfOptimizer
from NiaPy.task.task import StoppingTask, OptimizationType

# prepare instance of StoppingTask with the following parameters:
# D - dimension of problem set to 10
# nFES - number of function evaluations set top 1000
# benchmark - benchmark function set to sphere
task = StoppingTask(D=10, nFES=1000, benchmark='sphere')
algorithm = GreyWolfOptimizer(NP=20)
best = algorithm.run(task=task)
print('%s -> %s' % (best[0], best[1]))

# plot a convergence graph
task.plot()
```
